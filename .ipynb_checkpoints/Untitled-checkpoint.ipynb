```python
## formula for calculating time:
## get all the time stamp of the files
## find the most common "date"
## in the most common date find the initial and final timestamp
## get the initial time stamp's starting time.
## subtract the starting time from the final timestamp
## divide that difference with the number of scans found with that most common date
## this will give you the time per scan
```

```python
import pandas as pd
from datetime import datetime,timedelta
import matplotlib.pyplot as plt
import numpy as np
### for scan selection:
# df=pd.read_csv('ibio_session_df_processing_files_time_stamp_1_1.csv')
df=pd.read_csv('SAH_session_df_processing_files_time_stamp_1_2.csv')
df.columns
```

```python
# df['NIFTI_LOCATION_TIMESTAMP']
project='SAH'
timestamps=['pdf_TIMESTAMP',
       '_4DL_seg.nii.gz_TIMESTAMP',
       '_normalized.nii.gz_csf_10.nii.gz_TIMESTAMP','_4DL_seg_total.nii.gz_TIMESTAMP']
# program_intiation_datetime=['2/23/2024, 9:38:06 AM','2/23/2024, 9:38:06 AM','2/23/2024, 9:38:06 AM','2/23/2024, 9:38:06 AM'] # #
program_intiation_datetime=['5/31/2024, 2:53:24 PM' ,'3/29/2024, 1:44:42 PM','4/18/2024, 8:48:01 AM','4/19/2024, 7:02:10 AM']
initiation_date='7/6/2023, 1:14:59 PM' #'2/23/2024, 9:38:06 AM'
delta = timedelta(hours=5, minutes=30)
date_format = "%m/%d/%Y, %I:%M:%S %p" #"%m/%d/%Y, %H:%M:%S"
# date_str=program_intiation_datetime[timestamp_col_id] #'5/31/2024, 2:53:24' #'3/18/2024, 12:35:42'
initiation_date_datetime = datetime.strptime(initiation_date, date_format)
initiation_date_datetime=initiation_date_datetime.replace(tzinfo=None) +delta

duration_each_step=['infarct_biomarker','preprocessing','segmentation','postprocessing']
dictionary_step={}
# dictionary_step[duration_each_step[0]]=678
# dictionary_step[duration_each_step[1]]=678
# dictionary_step
num_days=120
start_date=initiation_date_datetime
def find_max(numlist):
    return np.argmax(numlist)
this_list=[]
for i in range(num_days):
    # Add the current date to the list
    # date_list.append(start_date + timedelta(days=i))
    this_datetime=start_date + timedelta(days=i)
    # print(this_datetime)

    if len(duration_each_step) == len(program_intiation_datetime) and len(timestamps)== len(program_intiation_datetime) :
        
        for timestamp_col_id in range(len(duration_each_step)):
            # timestamp_col_id=3
            # print('Day number::{}'.format(i))
            # print('THIS IS FOR  ::{}'.format(duration_each_step[timestamp_col_id]))
            df['datetime_column'] = pd.to_datetime(df[timestamps[timestamp_col_id]], errors='coerce').dt.tz_localize(None)
            
            # this_timestamp=df['datetime_column']
            this_timestamp=df['datetime_column'].fillna(pd.NaT)
            this_timestamp=this_timestamp.dropna()
            diff_from_initial_time=(this_timestamp-this_datetime).dt.total_seconds() / 3600
            # print(diff_from_initial_time)
            # print(df['datetime_diff_from_initialtime'])
            df['datetime_diff_from_initialtime']=(df['datetime_column']-this_datetime).dt.total_seconds() / 3600
            # min_row=df[df['datetime_diff_from_initialtime']]
            # min_value_index = df['datetime_diff_from_initialtime'].dropna().idxmin()
            # min_value_row = df.loc[min_value_index]
            filtered_values = df['datetime_diff_from_initialtime'].dropna()[df['datetime_diff_from_initialtime'] > 0]

            # Find the index of the minimum value greater than zero
            if len(filtered_values)>0:
                min_value_index = filtered_values.idxmin()
                min_value_row = df.loc[min_value_index]
                xx=diff_from_initial_time[(diff_from_initial_time<=24) & (diff_from_initial_time>0)]
                if len(xx)>0:
#                     #####################
                    date_str_1=min_value_row[timestamps[timestamp_col_id]]
                    delta = timedelta(hours=5, minutes=30)
                    date_with_tz = datetime.fromisoformat(date_str_1)

                    # Remove the timezone information
                    date_without_tz = date_with_tz.replace(tzinfo=None)
                    # date_format = "%m/%d/%Y, %I:%M:%S %p" #"%m/%d/%Y, %H:%M:%S"
                    # date_str=program_intiation_datetime[timestamp_col_id] #'5/31/2024, 2:53:24' #'3/18/2024, 12:35:42'
                    # initiation_date_datetime_1 = datetime.strptime(date_str_1, date_format)
                    # initiation_date_datetime_1=initiation_date_datetime.replace(tzinfo=None) +delta
                    df['datetime_diff_from_initialtime_1']=(df['datetime_column']-date_without_tz).dt.total_seconds() / 3600
                    diff_from_initial_time_1=df['datetime_diff_from_initialtime_1']
                    # min_row=df[df['datetime_diff_from_initialtime']]
                    # min_value_index = df['datetime_diff_from_initialtime'].dropna().idxmin()
                    # min_value_row = df.loc[min_value_index]
                    filtered_values_1 = df['datetime_diff_from_initialtime_1'].dropna()[df['datetime_diff_from_initialtime_1'] > 0]

                    # Find the index of the minimum value greater than zero
                    if len(filtered_values_1)>0:
                        # min_value_index = filtered_values_1.idxmin()
                        # min_value_row = df.loc[min_value_index]
                        xx_1=diff_from_initial_time_1[(diff_from_initial_time_1<=24) & (diff_from_initial_time_1>0)]

                    ######################
                        this_list.append([str(min_value_row['label']),duration_each_step[timestamp_col_id],len(xx),len(xx_1),min_value_row[timestamps[timestamp_col_id]],this_datetime])
                # print(len(xx))
        #     this_list.append([duration_each_step[timestamp_col_id],xx])
# this_list_np=np.array(this_list)
# print(this_list)
this_list_df=pd.DataFrame(this_list)
this_list_df.columns=['SESSION_NAME','STEP_NAME','NUM_FILES_PRODUCED','NUM_FILES_PRODUCED_ORIG','THIS_TIME','STARTINGTIME']
this_list_df = this_list_df.sort_values(by='STEP_NAME')
this_list_df.to_csv(project+'date_maxoutput_stepwise.csv',index=False)
#         idmax=find_max(this_list_np[:,1])
#         print(this_list_np[idmax,:])
 
```
